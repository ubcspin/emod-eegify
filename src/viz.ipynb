```python
import os
import utils
import numpy as np
import pandas as pd

from config import EXP_PARAMS
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
SUBJECT_IDS = ['p02', 'p04', 'p05', 'p06', 'p07', 'p08', 'p09', 'p10', 'p12', 'p13', 'p15', 'p17', 'p19', 'p20', 'p22', 'p23']

FEATURE_TYPES = ['alpha', 'beta', 'delta', 'theta', 'gamma']
LABEL_TYPES = [
    'pos',
    'angle',
    'acc',
    # 'cw_list',
    # 'cw_mode',
]



INPUT_DIR = 'RESULTS'
INPUT_PICKLE_NAME = 'results.pk'
```

```python
participant_results = {}
for i in range(100, 101):
    for window_size in EXP_PARAMS['WINDOW_SIZE']:
        for subject_id in SUBJECT_IDS:
            iter_ = subject_id + "_" + str(i) + '_' + str(window_size) + 'ms_hc_cw_' + INPUT_PICKLE_NAME
            input_pickle_file_path = os.path.join(INPUT_DIR, iter_)
            participant_results[subject_id + "_" + str(i) + '_' + str(window_size)] = utils.load_pickle(pickled_file_path=input_pickle_file_path)
            participant_results[subject_id + "_" + str(i) + '_' + str(window_size)]['window'] = str(window_size)
            participant_results[subject_id + "_" + str(i) + '_' + str(window_size)]['iter'] = i
```

```python
dataset = {'pnum': [], 'window': [], 'dir': [], 'cw' : [], 'iter': []}

score_set_columns = ['pnum', 'window', 'hc-type', 'estimator', 'min_f1', 'mean_f1', 'max_f1', 'std_f1', 'min_precision', 'mean_precision', 'max_precision', 'std_precision', 'min_recall', 'mean_recall', 'max_recall', 'std_recall']
score_set = []#{'pnum': [], 'window': [], 'hc-label' : [], 'estimator':[],	'min_score':[],	'mean_score':[], 'max_score':[], 'std_score':[] }

for key in participant_results.keys():
    result = participant_results[key]
    
    #pos = result['y_hc_pos'][:, 1]
    dir = result['y_hc_angle'][:, 1]
    #acc = result['y_hc_acc'][:, 1]

    cw = result['y_hc_angle'][:, 0]


    pnum = [result['pnum']] * len(cw)
    window = [result['window']] * len(cw)
    iter = [result['iter']] * len(cw)

    dataset['pnum'].extend(pnum)
    dataset['window'].extend(window)
    dataset['iter'].extend(iter)
    
    #dataset['pos'].extend(pos)
    dataset['dir'].extend(dir)
    #dataset['acc'].extend(acc)

    dataset['cw'].extend(cw)

    #score_pos = [result['pnum'], result['window'], 'hc_pos'] + list(result['scores_hc_pos'].to_numpy().squeeze())
    score_dir = [result['pnum'], result['window'], 'hc_dir'] + list(result['scores_hc_angle'].to_numpy().squeeze())
    #score_acc = [result['pnum'], result['window'], 'hc_acc'] + list(result['scores_hc_acc'].to_numpy().squeeze())

    #score_set.append(score_pos)
    score_set.append(score_dir)
    #score_set.append(score_acc)


df_scores = pd.DataFrame(score_set,
                   columns=score_set_columns)
df_dataset = pd.DataFrame(data=dataset)
```

```python
def precision(y_true: np.ndarray, y_pred: np.ndarray):
    sum_intersection = 0
    sum_prediction_and_ancestors = 0
    for ground_truth, prediction in zip(y_true, y_pred):
        ground_truth_set = set(ground_truth)
        prediction_set = set(prediction)
        sum_intersection = sum_intersection + len(
            ground_truth_set.intersection(prediction_set)
        )
        sum_prediction_and_ancestors = sum_prediction_and_ancestors + len(
            prediction_set
        )
    precision = sum_intersection / (sum_prediction_and_ancestors)
    return precision


def recall(y_true: np.ndarray, y_pred: np.ndarray):
    sum_intersection = 0
    sum_prediction_and_ancestors = 0
    for ground_truth, prediction in zip(y_true, y_pred):
        ground_truth_set = set(ground_truth)
        prediction_set = set(prediction)
        sum_intersection = sum_intersection + len(
            ground_truth_set.intersection(prediction_set)
        )
        sum_prediction_and_ancestors = sum_prediction_and_ancestors + len(
            ground_truth_set
        )
    recall = sum_intersection / (sum_prediction_and_ancestors)
    return recall

def accuracy(y_true: np.ndarray, y_pred: np.ndarray):
    score = 0
    for ground_truth, prediction in zip(y_true, y_pred):
        score += (ground_truth[0] == prediction[0]) and (ground_truth[0] == prediction[0])
    score /= len(y_true)
    return score

def f1(y_true: np.ndarray, y_pred: np.ndarray):
    prec = precision(y_true, y_pred)
    rec = recall(y_true, y_pred)
    f1 = 2 * prec * rec / (prec + rec + 1e-15)
    return f1



def row(pnum, window, scores):

    mean_f1, mean_prec, mean_recall, mean_acc = scores.mean(axis=0)
    std_f1, std_prec, std_recall, std_acc = scores.std(axis=0)
    d = {
        'pnum': pnum,
        'window': window,

        
        'chance_f1': mean_f1,
        'std_chance_f1': std_f1,

        'chance_precision': mean_prec,
        'std_chance_precision': std_prec,


        'chance_recall': mean_recall,
        'std_chance_recall': std_recall,

        'chance_accuracy': mean_acc,
        'std_chance_accuracy': std_acc
        }
    return pd.Series({**d})


metric = lambda yt,yp: [f1(yt,yp), precision(yt,yp), recall(yt,yp)]



columns = ['pnum', 'window', 'chance_f1', 'std_chance_f1','chance_precision', 'std_chance_precision', 'chance_recall', 'std_chance_recall', 'chance_accuracy', 'std_chance_accuracy']
rows = []
for pnum in SUBJECT_IDS:
    for window in ['500', '5000']:
        n = 1000

        labelset = df_dataset[(df_dataset['pnum'] == pnum) & (df_dataset['window'] == window)][['cw', 'dir']]
        labels = df_dataset[(df_dataset['pnum'] == pnum) & (df_dataset['window'] == window)][['cw', 'dir']].values

        cw_unique = labelset.cw.unique() # number of unique words
        dir_unique = ['0','1','2'] # number of unique directions

        #print(labelset.cw.value_counts())
        metrics = np.zeros((n,4))

        cw_pred = np.random.choice(cw_unique, size=(labels.shape[0], n))
        dir_pred = np.random.choice(dir_unique, size=(labels.shape[0], n))

        # cw_pred_2 = np.random.choice(cw_unique, size=(labels.shape[0], n))
        # dir_pred_2 = np.random.choice(dir_unique, size=(labels.shape[0], n))
    
    

        for i in range(n):
            pred = np.vstack([cw_pred[:,i],dir_pred[:,i]]).T

            prec, rec = precision(labels,pred), recall(labels,pred)
            acc = accuracy(labels,pred)
            fscore = 2 * prec * rec / (prec + rec + 1e-15)
            metrics[i] = [fscore, prec, rec, acc]
            #metrics[i] = metric(labels, pred)
            #print(metrics[i])

        rows.append((row(pnum, window, metrics)))

df_chances = pd.concat(rows, axis=1).T.sort_values(['pnum'], ascending=False)
df_chances

# df_chances.to_csv('chance_values.csv', index=False)
```
